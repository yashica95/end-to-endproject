## Overview: 
This project uses the Brazilian E-Commerce Public Dataset published by Olist on Kaggle to demonstrate the scoping and implementation of a real-world data science project. The dataset's description on Kaggle states:

### Context: 
This dataset was generously provided by Olist, the largest department store in Brazilian marketplaces. Olist connects small businesses from all over Brazil to channels without hassle and with a single contract. Those merchants are able to sell their products through the Olist Store and ship them directly to the customers using Olist logistics partners. 

A seller join Olist through a marketing and sales funnel that was made public at this dataset. Description of steps:

Sign-up at a landing page.
Get contacted by a Sales development Representative (SDR), confirm some information and schedule a consultancy.
Consultancy is made by a Sales Representative (SR). The SR may close the deal (lead sing up) or lose the deal (led leaves without sign in)
Lead becomes a seller and starts building his catalog on Olist.
His products are published on marketplaces and ready to sell!

Attention
A seller MQL might come from multiple sources (he might subscribe on two different landing pages, for instance)

### Data: 
The dataset has information of 8000 Marketing Qualified Leads (MQLs) from Jun 2017 - May 2018. The Olist's team observed a 10.5% conversion rate to close the leads i.e. out of 8000 potential leads, 842 sellers were approved to sell on the marketplace. The data contains information about how the lead was acquired: landing page and channel acquisition and also information about when the deal was closed, by whom and additional information about the seller. 

Its features allows viewing a sales process from multiple dimensions: lead category, catalog size, behaviour profile, etc. This is real data, it has been anonymized and sampled from the original dataset.
More information can be found here: https://www.kaggle.com/olistbr/marketing-funnel-olist

### Goal:

The goal is to find a causal inference during the marketing journey and find out if there a factors that might cause a lead to be closed such as the channel acquisition of the seller. Since 10 different sources were used by the Olist to acquire the leads, marketing team wants to understand if the source of acquisition would be an important factor in determining the conv rate of the leads. 

Using the results of this exercise, marketing team should be able to:

* Determine the cause of a lead being closed.

## EDA

Analysis the distribution of variables such as acquisition source and landing page to understand the relation with whether the deal was closed or not 

### Source of Acquisition

```python
#Understanding the distribution of leads by acquisition source
sns.displot( data= final, y="origin", hue="closed", multiple="dodge", palette="rocket")
plt.show();
```

Clearly, we need to combine sources like other, Other, unknown etc into a single category and compare the performance across different channels especially paid_search, social and organic_search

```python
def sources(x):
    if x in ["Other", "other", "other_publicities"]:
        return "Other"
    else:
        return x

final["origin"] = final["origin"].apply(lambda x: sources(x))
```

```python
sources_report = final.pivot_table(index=['origin'],columns=['closed'],values='mql_id',aggfunc='count',fill_value=0).reset_index()
sources_report.columns = ["Origin", "Lost Leads", "Converted Leads"]
sources_report["Total Leads"] = sources_report["Lost Leads"] + sources_report["Converted Leads"]
sources_report["Conversion Rate (%)"] = round(sources_report["Converted Leads"]*100/ sources_report["Total Leads"],2)

#Understanding the distribution of leads by acquisition source
sns.barplot(x=sources_report["Origin"], y=sources_report["Conversion Rate (%)"], palette="rocket");
```

From the graph we can draw insights that paid search has the highest conversion rate of ~12% followed by organic search. Social, display & email acquisition sources have the lowest conversion rate. Therefore it seems that acquisition source plays a key role in determining whether a lead would convert as a seller or not. 

### Exploring the effect of  Landing Page

```python
final['landing_page_id'].value_counts()
```

Apparently, 495 different landing pages were used to signup the leads during the given time period but the top two landing pages being "22c29808c4f815213303f8933030604c" and "b76ef37428e6799c421989521c0e5077". To consider A/B experiment of landing pages, we can consider these two landing pages and see if one page was better than the other. 

```python
top_landing_pages = final[(final["landing_page_id"]=="22c29808c4f815213303f8933030604c") |(final["landing_page_id"]=="b76ef37428e6799c421989521c0e5077") ]
top_landing_pages["landing_page_id"].value_counts()
```

```python
landing_pages = top_landing_pages.pivot_table(index=['landing_page_id'],columns=['closed'],values='mql_id',aggfunc='count',fill_value=0).reset_index()
landing_pages.columns = ["Landing Page", "Lost Leads", "Converted Leads"]
landing_pages["Total Leads"] = landing_pages["Lost Leads"] + landing_pages["Converted Leads"]
landing_pages["Conversion Rate (%)"] = round(landing_pages["Converted Leads"]*100/ landing_pages["Total Leads"],2)

#Understanding the distribution of leads by acquisition source
sns.barplot(x=landing_pages["Landing Page"], y=sources_report["Conversion Rate (%)"], palette="rocket");
```

We can clearly see that one of the landing page has higher conversion rate compared to other landing page. We can treat the "landing_page_id" as a treatment i.e. assume that marketing team was trying to experiment with different landing pages and narrowed it down to these two landing pages. Does the landing page affect if the seller will close the deal or not or is it just coincidence?
For the sake of it, we will assume that leads who signed up via "22c29808c4f815213303f8933030604c" landing page as treatment group and those who signed up via "b76ef37428e6799c421989521c0e5077" as control group 

```python
def treatment(x):
    if x =="22c29808c4f815213303f8933030604c":
        return 1
    elif x=="b76ef37428e6799c421989521c0e5077":
        return 2
    else:
        return 0
final["treatment"] =  final["landing_page_id"].apply(lambda x: treatment(x))
final.head(2)
```

## CausalML 

```python
#installing the package 
#!pip install causalml
import causalml
```

```python
from causalml.inference.meta import XGBTRegressor, MLPTRegressor
from causalml.inference.meta import BaseSRegressor, BaseTRegressor, BaseRRegressor, BaseXRegressor, LRSRegressor # method for meta learners
from causalml.match import NearestNeighborMatch, create_table_one, MatchOptimizer ## Used for propensity score match
from causalml.propensity import ElasticNetPropensityModel  ## Used for propensity score match
from causalml.features import load_data
from causalml.features import LabelEncoder, OneHotEncoder
from causalml.metrics.sensitivity import Sensitivity ## used for sensitivity analysis
from causalml.dataset import *
from causalml.metrics import *
```

```python
#dummifying origin column for casual ML
df = pd.get_dummies(final, columns=['origin'])
```

```python
OUTCOME = df["closed"] #the column to be used as the outcome i.e whether a deal would be closed or not 
TREATMENT = df["treatment"]#which landing page was shown to the seller would be the treatment 
FEATURES = df.loc[:, "origin_Other":]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error as mse
```

## Comparing different models

```python
# Ready-to-use T-Learner using XGB
learner_t = XGBTRegressor()
ate_t = learner_t.estimate_ate(X=FEATURES, treatment=TREATMENT, y=OUTCOME)
print('Using the ready-to-use XGBTRegressor class')
print(ate_t)

# Calling the Base Learner class and feeding in XGB
learner_t = BaseTRegressor(learner=XGBRegressor())
ate_t = learner_t.estimate_ate(X=FEATURES, treatment=TREATMENT, y=OUTCOME)
print('\nUsing the BaseTRegressor class and using XGB (same result):')
print(ate_t)

# Calling the Base Learner class and feeding in LinearRegression
learner_t = BaseTRegressor(learner=LinearRegression())
ate_t = learner_t.estimate_ate(X=FEATURES, treatment=TREATMENT, y=OUTCOME)
print('\nUsing the BaseTRegressor class and using Linear Regression (different result):')
print(ate_t)
```

```python
s_learner = BaseSRegressor(LGBMRegressor())
s_ate = s_learner.estimate_ate(FEATURES, TREATMENT, OUTCOME)[0]
s_ite = s_learner.fit_predict(FEATURES, TREATMENT, OUTCOME)
```

## Using UpliftTreeClassifier for One control and Multiple Treatment Groups

```python
from causalml.inference.tree import UpliftTreeClassifier, UpliftRandomForestClassifier
from causalml.dataset import make_uplift_classification
from causalml.inference.tree import uplift_tree_string, uplift_tree_plot
```

```python
def treatment_key(x):
    if x==1:
        return "treatment A"
    elif x==2:
        return "treatment B"
    else:
        return "control"
df["treatment_key"] =  df["treatment"].apply(lambda x: treatment_key(x))
```

```python
uplift_tree = UpliftTreeClassifier(control_name='control')

TREATMENT = df["treatment_key"]

uplift_tree.fit(X=FEATURES,
                treatment=TREATMENT,
                y=OUTCOME)
```

```python
df.pivot_table(values='closed',
               index='treatment_key',
               aggfunc=[np.mean, np.size],
               margins=True)
```

```python
# Split data to training and testing samples for model validation (next section)
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=111)

# Train uplift tree
uplift_model = UpliftTreeClassifier(max_depth = 3, min_samples_leaf = 200, min_samples_treatment = 50, n_reg = 100, evaluationFunction='KL', control_name='control')

uplift_model.fit(X=FEATURES,
                treatment=TREATMENT,
                y=OUTCOME)
```

```python
# Plot uplift tree
# The uplift score represents the best uplift score among all treatment effects
from IPython.display import Image
graph = uplift_tree_plot(uplift_model.fitted_uplift_tree,FEATURES.columns.values)
Image(graph.create_png())
```

# Conclusion

According to the UpliftClassifier model, it seems that knowing if the seller was acquired through a paid search source such as Google Ads/ Facebook Ads etc is the most important feature to determine if the deal will be closed. If the seller was acquired through a paid channel, landing page A i.e. 22c29808c4f815213303f8933030604c would be the most effective page whereas if the lead isn't acquired via paid channel, the page 22c29808c4f815213303f8933030604c and b76ef37428e6799c421989521c0e5077 can be used interchangeably. 

Therefore, it is advised to the marketing team to use 22c29808c4f815213303f8933030604c as a landing page for the campaigns being run on Google Ads, facebook ads etc paid channels as it has higher uplift score than other landing pages.

Also it can be seen from the graph that paid_search channels have a lower uplift score, compared to leads that weren't acquired through paid channels - although it isn't clear which channel is better.
